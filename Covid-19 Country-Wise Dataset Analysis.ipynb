### Exploratory data analysis

#### Objective 1: Finding the top ten countries with the highest confirmed COVID-19 cases.

```python
# Grouping countries by confirmed cases
conf_cases = pd.DataFrame(df.groupby('Country/Region').sum()['Confirmed'])

# Sorting the dataframe in descending order
conf_cases.sort_values(by=['Confirmed'], inplace=True, ascending=False)

# Top 10 countries by confirmed cases
conf_cases[:10]
```

#### Objective 2: Finding the top ten countries with the highest number of deaths from COVID-19.

```python
# Grouping countries by number of deaths
deaths = pd.DataFrame(df.groupby('Country/Region').sum()['Deaths'])

# Sorting the dataframe in descending order
deaths.sort_values(by=['Deaths'], inplace=True, ascending=False)

# Top 10 countries by number of deaths
deaths[:10]
```

#### Objective 3: Finding the top ten countries with the highest number of COVID-19 recovered cases.

#### Objective 4: Finding the top ten countries with the highest number of COVID-19 active cases

```python
# Grouping countries by number of active cases
con_cases = pd.DataFrame(df.groupby('Country/Region').sum()['Active'])

# Sorting the dataframe in descending order
con_cases.sort_values(by=['Active'], inplace=True, ascending=False)

# Top 10 countries by number active cases
con_cases[:10]
```

#### Objective 5: Finding the top ten countries with the highest number of confirmed cases in relation to the previous week.

```python
# Grouping countries by confirmed cases in relation to the prevoius week
conf_cases = pd.DataFrame(df.groupby('Country/Region',).sum()[{'Confirmed','Confirmed last week'}])

# Sorting the dataframe in descending order
conf_cases.sort_values(by=['Confirmed last week'], inplace=True, ascending=False)

# Top 10 countries by confirmed cases relation to the prevoius week
conf_cases[:10]
```

#### Objective 6: Finding the top ten countries with the highest number of death cases in relation to the percentage deaths.

```python
# Grouping countries by death cases in relation to the percentage deaths.
conf_cases = pd.DataFrame(df.groupby('Country/Region',).sum()[{'Deaths','Deaths / 100 Cases'}])

# Sorting the dataframe in descending order
conf_cases.sort_values(by=['Deaths / 100 Cases'], inplace=True, ascending=False)

# Top 10 countries by death cases in relation to the percentage deaths.
conf_cases[:10]
```

#### Objective 7: Finding the top ten countries with the highest number of recovered cases in relation to the percentage recovered cases.

```python
# Grouping countries by recovered cases in relation to the percentage recoveries.
conf_cases = pd.DataFrame(df.groupby('Country/Region',).sum()[{'Recovered','Recovered / 100 Cases'}])

# Sorting the dataframe in descending order
conf_cases.sort_values(by=['Recovered / 100 Cases'], inplace=True, ascending=False)

# Top 10 countries by recovered cases in relation to the percentage recoveries.
conf_cases[:10]
```

#### Objective 8: Finding the top ten countries with the highest number of  deaths/100 recovered cases in relation to the death cases

```python
# Grouping countries by recovered cases in relation to the percentage deaths.
conf_cases = pd.DataFrame(df.groupby('Country/Region',).sum()[{'Recovered','Recovered / 100 Cases','Deaths / 100 Recovered'}])

# Sorting the dataframe in descending order
conf_cases.sort_values(by=['Deaths / 100 Recovered'], inplace=True, ascending=False)

# Top 10 countries by recovered cases in relation to the percentage recoveries.
conf_cases[:10]
```

#### Objective 9: Finding the top ten countries with the highest number 1 week change cases.

```python
# Grouping countries by number 1 week change cases
conf_cases = pd.DataFrame(df.groupby('Country/Region').sum()['1 week change'])

# Sorting the dataframe in descending order
conf_cases.sort_values(by=['1 week change'], inplace=True, ascending=False)

# Top 10 countries by number 1 week change cases
conf_cases[:10]
```

#### Objective 10: Finding the top ten countries with the highest number of 1 week % change cases in relation to the 1 week change.

```python
# Grouping countries by 1 week % increase cases in relation to the 1 week change.
conf_cases = pd.DataFrame(df.groupby('Country/Region',).sum()[{'1 week change','1 week % increase'}])

# Sorting the dataframe in descending order
conf_cases.sort_values(by=['1 week % increase'], inplace=True, ascending=False)

# Top 10 countries by 1 week % increase cases in relation to the 1 week change.
conf_cases[:10]
```
